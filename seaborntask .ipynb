```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Create some sample data
data = np.random.randn(100, 4)

# Convert the NumPy array to a pandas DataFrame
df = pd.DataFrame(data, columns=['A', 'B', 'C', 'D'])

# Create a histogram
sns.histplot(df['A'], kde=True)
plt.title('Hameedas Histogram Example')
plt.show()

# Create a boxplot
sns.boxplot(data=df)
plt.title('Hameedas Boxplot Example')
plt.show()

# Create a pairplot (includes histograms and scatter plots)
sns.pairplot(df)
plt.show()
```
