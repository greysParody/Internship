```python
import matplotlib.pyplot as plt
import numpy as np

# Create some sample data
x = np.arange(10)
y1 = np.random.rand(10)
y2 = np.random.rand(10)

# Create a line plot
plt.plot(x, y1, label='Line 1')
#plt.plot(x, y2, label='Line 2')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Hameeda Line Plot')
plt.legend()
plt.show()

# Create a scatter plot
plt.scatter(x, y1, label='Scatter 1')
plt.scatter(x, y2, label='Scatter 2')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Hameedas Scatter Plot')
plt.legend()
plt.show()
```
