```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns  

# Task 1
# Provided data
months = list(range(1, 13))
total_profit = [211000, 183300, 224700, 222700, 209600, 201400, 
                295500, 361400, 234000, 266700, 412800, 300200]

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(months, total_profit, marker='o', linestyle='-', color='b', label='Total Profit')
plt.title('Total Profit of Each Month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.xticks(months)  # Ensure all months are shown on the x-axis
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns  

# Extracting data for bathing soap and face wash
bathingsoap = [9200, 6100, 9550, 8870, 7760, 7490, 8980, 9960, 8100, 10300, 13300, 14400]
facewash = [1500, 1200, 1340, 1130, 1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760]

# Plotting subplot
plt.figure(figsize=(12, 6))

# Subplot 1 for Bathing Soap
plt.subplot(1, 2, 1)
plt.plot(months, bathingsoap, marker='o', linestyle='-', color='g')
plt.title('Bathing Soap Sales')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.xticks(months)
plt.grid(True)

# Subplot 2 for Face Wash
plt.subplot(1, 2, 2)
plt.plot(months, facewash, marker='o', linestyle='-', color='r')
plt.title('Face Wash Sales')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.xticks(months)
plt.grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```
